```python
from bs4 import BeautifulSoup
```

```python
url="https://companiesmarketcap.com/tesla/revenue"
```

```python
import requests
response=requests.get(url)
```

```python
response
```

```python
soup=BeautifulSoup(response.text,"html")
```

```python
table=soup.find("table",class_="table")
```

```python
header=table.find_all("th")
```

```python
header
```

```python
rows=table.find_all("tr")
```

```python
rows
```

```python
Revenue=[]
year=[]
for r in rows[1:]:
    data=r.find_all("td")
    Revenue.append(data[1].text)
    year.append(data[0].text)
    
```

```python
print(Revenue)
```

```python
print(year)
```

```python
import pandas as pd
data=pd.DataFrame({"year=":year,"Revenue":Revenue})
data
```

```python
data.tail()
```

```python
data
```

```python
data["year="] = data['year='].str.replace('(TTM)',"")
```

```python
data
```

```python
data["Revenue"] = data['Revenue'].str.replace('B',"")
```

```python
data
```

```python
data["Revenue"] = data['Revenue'].str.replace('$',"")
```

```python
data
```

```python
data['Revenue'] = pd.to_numeric(data['Revenue'])
```

```python
data
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(data['year='], data['Revenue'], marker='o', linestyle='-')
plt.title('Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()
```

```python
data['year='] = pd.to_datetime(data['year='])
```

```python
soup=BeautifulSoup(response.text,"html")
```

```python
table=soup.find('table',class_="svelte-l35b6s")
```

```python
table
```

```python
header=table.find_all("th")
```

```python
header
```

```python
rows=table.find_all("tr")
```

```python
rows
```

```python
Revenue=[]
year=[]
for r in rows[1:]:
    data=r.find_all("td")
    Revenue.append(data[1].text)
    year.append(data[0].text)
```

```python
print(Revenue)
```

```python
print(year)
```

```python
import pandas as pd
data=pd.DataFrame({"year":year,"Revenue":Revenue})
data
```

```python
data["Revenue"] = data['Revenue'].str.replace('B',"")
data["Revenue"] = data['Revenue'].str.replace('M',"")
```

```python
data
```

```python
data['Revenue'] = pd.to_numeric(data['Revenue'])
```

```python
data
```

```python
data['year'] = pd.to_datetime(data['year'])
data=data.iloc[:20,:]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(data['year'], data['Revenue'], marker='o', linestyle='-')
plt.title('Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()
```

```python
def make_graph(data):
    import matplotlib.pyplot as plt

    plt.figure(figsize=(10, 6))
    plt.plot(data['year'], data['Revenue'], marker='o', linestyle='-')
    plt.title('Revenue Over Time')
    plt.xlabel('Date')
    plt.ylabel('Revenue')
    plt.grid(True)
    plt.show()
```

```python
make_graph(data)
```
